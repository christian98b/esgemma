```python
from datasets import load_dataset
```

```python
import pandas as pd
repo_name = 'chris7374/esg-net-zero'
revision = '100_per_class_v3'
dataset = load_dataset(repo_name, revision=revision)
pd_dataset = dataset['train'].to_pandas()
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("google/gemma-2-2b-it")
```

```python
import matplotlib.pyplot as plt
token_counts = [len(tokenizer.encode(x)) for x in pd_dataset.custom_text]
# plot the token counts
a = plt.hist(token_counts, bins=30)
```
